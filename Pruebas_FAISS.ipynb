```python
from langchain_community.llms import Ollama

llm = Ollama(model="llama3",
             temperature=0.0) # base_url = 'http://localhost:11434'

llm
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "Tu eres un asistente especializado en ofertas del empleo público para ayudarte cuentas con una base de datos proporcionada via RAG con las ofertas recientemente publicadas, se conciso, es importante, no inventes datos, ni rellenes con datos no proporcionados en el contexto"),
    ("user", "{query}")
])
```

```python
llm_app = prompt | llm

llm_app
```

```python
response = llm_app.invoke({"query": "Me puedes decir ofertas de empleo publico relaccionadas con la informática?"})

print(response)
```

```python
import pandas as pd
from langchain_core.documents.base import Document
```

```python
import pandas as pd
df = pd.read_csv('/content/recuperacion_informacion_modelos_lenguaje/tfm/ObtencionDatos/datos/csv_boes_oferta_publica.csv',
                 sep='|')
```

```python
df.head()
```

```python
# Lista para almacenar los documentos
documentos = []

# Iterar sobre cada fila del DataFrame
for index, row in df.iterrows():
    # Crear un objeto Document
    doc = Document(
        page_content=row['texto'],   # El contenido principal del documento
        metadata={
            'titulo': row['titulo'],
            'url': row['url']
        }
    )
    # Añadir el Documento a la lista
    documentos.append(doc)
```

```python
documentos[0]
```

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings_llm = OllamaEmbeddings(model="llama3") # base_url = 'http://localhost:11434'
```

```python
embeddings = embeddings_llm.embed_query("Cómo estás tú?")

embeddings[:5]
```

```python
type(embeddings), len(embeddings)
```

## 2.3 Build Index in Vector Store

-
pip install faiss-cpu

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter()

documents = text_splitter.split_documents(documentos)

len(documents)
```

```python
from langchain_community.vectorstores import FAISS

vector_index = FAISS.from_documents(documents, embeddings_llm)

vector_index
```

```python
retriever = vector_index.as_retriever()

```

```python
relevant_docs = retriever.invoke({"input": "Salud"})

len(relevant_docs)
```

```python
relevant_docs
```

```python
for doc in relevant_docs:
    print(f"Title : {doc.metadata['titulo']}, Source: {doc.metadata['url']}")
```

## 2.4 Create Retrieval Chain


```python
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.documents import Document

prompt = ChatPromptTemplate.from_template("""
Contesta las siguiente pregunta basandote en el contexto y en tu conocimiento interno.
Otorga prioridad al contexto y si no estás seguro di que no sabes la respuesta.

Otorga prioridad al contexo y a la base de datos proporcionada por RAG con Ofertas de empleo público cuando lo requieras

<context>
{context}
</context>

Question: {input}
""")

#document_chain  = prompt | llm
document_chain = create_stuff_documents_chain(llm, prompt)

response = document_chain.invoke({
    "input": "Sabes de alguna oferta de empleo publico de informático",
    "context": [Document(page_content="Leer todas las ofertas de empleo público facilitadas por FAISS")]
})

print(response)
```

```python
from langchain.chains import create_retrieval_chain

retrieval_chain = create_retrieval_chain(retriever, document_chain)

# retrieval_chain = retriever | document_chain

chain = prompt | document_chain | llm

#retrieval_chain
```

```python
response = retrieval_chain.invoke({"input": "Sabes de alguna oferta de empleo publico en sanidad",
                                   "context": [Document(page_content="Leer todas las ofertas de empleo público facilitadas por FAISS")]})

type(response)
```

```python
response.keys()
```

```python
for doc in response["context"]:
    print(f"Title : {doc.metadata['titulo']}, Source: {doc.metadata['url']}")
```

```python
response = chain.invoke({
    "input": {
        "query": "Sabes de alguna oferta de empleo publico en sanidad"
    },
    "context": "Sabes usar RAG FAISS"
})
```
